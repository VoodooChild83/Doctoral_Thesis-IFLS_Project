### IFLS

```python
#Read in the Data frame from Stata: IFLS Data

Wage_Data_Prov = ( pd.read_stata(Path_IFLS+"/Wages_Educ_Provinces.dta", convert_categoricals=False) 
                     .rename(columns={'provmov':'Prov Code'})
                     .dropna() ) 

Wage_Data_Prov["Prov Code"] = ( Wage_Data_Prov["Prov Code"].astype(int,copy=False) )
```

## Import Shape Files

```python
Ind_Prov = GeoDataFrame.from_file(Master_path+"Shape Files with Indonesian BPS Codes/Prov_merge/Prov_merge.shp")
# Ind_Kab = GeoDataFrame.from_file(Master_path+"Shape Files with Indonesian BPS Codes/Kab_merge/Kab_merge.shp")
```

```python
Ind_Prov = ( Ind_Prov.merge(Wage_Data_Prov, how='left', left_on="Prov Code", right_on="Prov Code") )
```
